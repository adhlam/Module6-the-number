 **Performing the base ten subtraction 25 − 37 using two’s complement**



Instead of 25-37, we compute 25 + (-37)

Converting 25 into base 2 number

```python
bin(25)
```

Converting 37 into a base 2 number

```python
bin(37)
```

Now, to convert into -37, first 00100101 is flip (i.e. 11011010)and the 1 is added to it which is equals to 11011011. Then finally adding 11011011 and 0011001 gives the result 11110100 which is the two's-complement.

Since, the left most bit is 1, so the above result is negative number.

To find out the number, first the above result is flip(i.e.00001011) and add 1 to it which results to 1100.

```python
int('1100',2)
```
